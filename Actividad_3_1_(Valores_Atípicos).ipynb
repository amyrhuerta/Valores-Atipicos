```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Ventas_totales_sin_nulos.csv')
df.head()
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df.info()
```

## ventas_precios_constantes

```python
fig = plt.figure(figsize=(10, 6))
plt.boxplot(df["ventas_precios_constantes"]) 
plt.title("Outliers de ventas_precios_constantes")
plt.show()
```

```python
y=df["ventas_precios_constantes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
outliers= df[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
df_clean= df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_clean
```

```python
fig = plt.figure(figsize =(5, 3))
plt.boxplot(df_clean["ventas_precios_constantes"]) 
plt.title("Outliers de ventas_precios_constantes")
plt.show()
```

```python
df_clean.to_csv("data_clean_ventas_precios_constantes.csv")
from google.colab import files  
files.download("data_clean_ventas_precios_constantes.csv") 
```
